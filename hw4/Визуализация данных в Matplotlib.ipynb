```python
# Задание 2
t = np.linspace(0, 10, 51)
t
```

```python
f = np.array([np.cos(i) for i in t])
f
```

```python
plt.plot(t, f, color="green")
plt.title('График f(t)')
plt.xlabel("Значения t")
plt.ylabel("Значения f")
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

```python
# Задание 3
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_xlim([-5, 5])
ax1.set_title('График y1')
ax2.plot(x, y2)
ax2.set_title('График y2')
ax3.plot(x, y3)
ax3.set_title('График y3')
ax4.plot(x, y4)
ax4.set_title('График y4')
```

```python
cls0 = df.loc[df['Class'] == 0, ['V1']]
cls1 = df.loc[df['Class'] == 1, ['V1']]
```

```python
plt.hist(cls0['V1'], bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
plt.hist(cls1['V1'], bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.xlabel('V1')
plt.legend()
plt.show()
```
